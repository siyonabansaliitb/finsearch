```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_option_pricing(S0, K, r, sigma, T, num_paths, num_steps):
    dt = T / num_steps
    S = np.zeros((num_steps + 1, num_paths))
    S[0, :] = S0

    for t in range(1, num_steps + 1):
        z = np.random.randn(num_paths)
        S[t, :] = S[t - 1, :] * np.exp((r - 0.5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * z)

    option_payoffs = np.maximum(S[-1, :] - K, 0)

    option_price = np.exp(-r * T) * np.mean(option_payoffs)

    return option_price, S

# Example usage
if __name__ == "__main__":
    S0 = 100 
    K = 100   
    r = 0.05 
    sigma = 0.2  
    T = 1.0   
    num_paths = 10000  
    num_steps = 252 

    option_price, S = monte_carlo_option_pricing(S0, K, r, sigma, T, num_paths, num_steps)
    print("Monte Carlo option price:", option_price)

    plt.figure(figsize=(10, 6))
    for i in range(10):
        plt.plot(np.arange(num_steps + 1), S[:, i])
    plt.xlabel("Time Steps")
    plt.ylabel("Stock Price")
    plt.title("Simulated Stock Price Paths")
    plt.grid(True)
    plt.show()
```
